```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as py
```

```python
file = pd.read_excel('Data_Extract.xlsx')
file.head()
```

```python
file.rename(columns = {'Series Name' : "SNam", 'Series Code' : 'SCod', 'Country Name' : 'CNam', 'Country Code' : 'CCod'},
            inplace = True)
file.columns
```

```python
file.columns = file.columns.str.slice(0,4,1)
file.columns
```

```python
codes = ['NY.GDP.MKTP.CD', 'NY.GDP.MKTP.KD.ZG', 'NV.AGR.TOTL.KD.ZG', 'NV.IND.TOTL.KD.ZG', 'NV.IND.MANF.KD.ZG',
         'NV.SRV.TOTL.KD.ZG', 'NE.CON.TOTL.KD.ZG', 'NE.GDI.TOTL.KD.ZG', 'NE.EXP.GNFS.KD.ZG', 'NE.IMP.GNFS.KD.ZG',
         'NV.AGR.TOTL.ZS', 'NV.IND.TOTL.ZS', 'NV.SRV.TOTL.ZS', 'NE.CON.TOTL.ZS', 'NE.GDI.TOTL.ZS', 'NE.EXP.GNFS.ZS',
         'NE.IMP.GNFS.ZS', 'NY.GNP.PCAP.CD', 'NY.GNP.PCAP.PP.CD', 'SP.POP.TOTL', 'NY.GNS.ICTR.ZS', 'NY.ADJ.SVNG.GN.ZS', 
         'TX.VAL.MRCH.XD.WD', 'TM.VAL.MRCH.XD.WD', 'BX.TRF.PWKR.DT.GD.ZS', 'BN.CAB.XOKA.GD.ZS', 'BX.KLT.DINV.WD.GD.ZS',
         'FP.CPI.TOTL', 'TX.UVI.MRCH.XD.WD', 'TM.UVI.MRCH.XD.WD', 'TT.PRI.MRCH.XD.WD', 'SL.GDP.PCAP.EM.KD', 'SL.UEM.TOTL.ZS',
         'NV.AGR.EMPL.KD', 'NV.IND.EMPL.KD', 'NV.SRV.EMPL.KD']
```

```python
New = file[file.SCod.isin(codes)]
New.shape
```

```python
New.head()
```

```python
GAES = ['NY.GDP.MKTP.CD', 'NY.GDP.MKTP.KD.ZG', 'NV.AGR.TOTL.KD.ZG', 'NV.IND.TOTL.KD.ZG', 'NV.IND.MANF.KD.ZG',
        'NV.SRV.TOTL.KD.ZG', 'NE.CON.TOTL.KD.ZG', 'NE.GDI.TOTL.KD.ZG', 'NE.EXP.GNFS.KD.ZG', 'NE.IMP.GNFS.KD.ZG',
        'NV.AGR.TOTL.ZS', 'NV.IND.TOTL.ZS', 'NV.SRV.TOTL.ZS', 'NE.CON.TOTL.ZS', 'NE.GDI.TOTL.ZS', 'NE.EXP.GNFS.ZS',
        'NE.IMP.GNFS.ZS']
IAS = ['NY.GNP.PCAP.CD', 'NY.GNP.PCAP.PP.CD', 'SP.POP.TOTL', 'NY.GNS.ICTR.ZS', 'NY.ADJ.SVNG.GN.ZS',]
BOP = ['TX.VAL.MRCH.XD.WD', 'TM.VAL.MRCH.XD.WD', 'BX.TRF.PWKR.DT.GD.ZS', 'BN.CAB.XOKA.GD.ZS', 'BX.KLT.DINV.WD.GD.ZS']
PATOT = ['FP.CPI.TOTL', 'TX.UVI.MRCH.XD.WD', 'TM.UVI.MRCH.XD.WD', 'TT.PRI.MRCH.XD.WD']
LAP = ['SL.GDP.PCAP.EM.KD', 'SL.UEM.TOTL.ZS','NV.AGR.EMPL.KD', 'NV.IND.EMPL.KD', 'NV.SRV.EMPL.KD']
```

```python
New.loc[New['SCod'].isin(GAES), 'Type'] = 'Growth and Economic Structure'
New.loc[New['SCod'].isin(IAS), 'Type'] = 'Income and Savings'
New.loc[New['SCod'].isin(BOP), 'Type'] = 'Balance of payments'
New.loc[New['SCod'].isin(PATOT), 'Type'] = 'Prices and terms of trade'
New.loc[New['SCod'].isin(LAP), 'Type'] = 'Labor and productivity'
```
